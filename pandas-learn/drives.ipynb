```python

df = pd.read_csv('../data/filtered-drivedata/toshibaMG07ACA14TA.csv')
display(df)
```

```python
# visualize correlations

# general heatmap of the data
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.show()
```

```python
# find pairs of high correlation and low correlation
numeric_columns = df.select_dtypes(include=["number"]).columns.tolist()
print(numeric_columns)

corr_tuples = []

for i in range(len(numeric_columns)):
    for j in range(i):
        corr_value = df[numeric_columns[i]].corr(df[numeric_columns[j]], min_periods=3)
        if np.isnan(corr_value):
            continue
        corr_tuples.append((numeric_columns[j], numeric_columns[i], corr_value))

```

```python

corr_df = pd.DataFrame(corr_tuples, columns=['property1', 'property2', 'correlation']).sort_values(by="correlation")
corr_df.sort_values(by="correlation")

# corr_df.head(5)
corr_df.tail(15)
```

```python
failure_df = pd.concat([corr_df[corr_df['property1'] == 'failure'], corr_df[corr_df['property2'] == 'failure']], ignore_index=True).sort_values(by='correlation')
display(failure_df)

# this is inconclusive, the failure statistic itself is not strongly correlated with anything. 
# the failure flag is only set once even though the drive may be "failing", also a lot of noise exists. 
# failure can occur for any reason, not just a single one
```

```python

cond = models.apply(lambda x : x > 1000)

# models with a large amount of drives
multi_models = models[cond]
print(multi_models)
```

```python

folder = '../data/drivedata' 
```

```python

folder = '../data/drivedata' 

significant = { m : set() for m in multi_models.index }

for filename in os.listdir(folder):
    if filename == 'schema.csv':
        continue

    df_file = pd.read_csv(os.path.join(folder, filename))
    print('Read', filename)
    # display(df_file)
    # display(df_file[df_file['model'] == 'ST4000DM000'])
    
    numeric = df_file.drop(df_file.filter(regex='normalized$'), axis=1).select_dtypes(include='number').columns.tolist()
    print(numeric)

    # for each model, find significant stats
    for m in multi_models.index:
        df_model = df_file[df_file['model'] == m]
        # print("model:", m)
        # print(m)
        # display(df_model)
        # break
        for col in numeric:
            # display(df_model[col])
            if len(df_model[col].unique()) > 1 and col not in significant[m]:
                significant[m].add(col)
                print('model:', m, '| col:', col)
                

```

```python
significant_pairs = []

for key in significant.keys():
    print(len(significant[key]))
    for property in significant[key]:
        if 'smart' in property:
            significant_pairs.append((key, property))


df_significant = pd.DataFrame(significant_pairs)

```
